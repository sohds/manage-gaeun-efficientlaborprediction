```python
import pandas as pd

df1 = pd.read_excel("xlsx/0910/221017.xlsx")
mon_df = pd.DataFrame()
mon_df = mon_df.append(df1)

df2 = pd.read_excel("xlsx/0910/221018.xlsx")
tue_df = pd.DataFrame()
tue_df = tue_df.append(df2)

df3 = pd.read_excel("xlsx/0910/221019.xlsx")
wed_df = pd.DataFrame()
wed_df = wed_df.append(df3)

df4 = pd.read_excel("xlsx/0910/221013.xlsx")
thur_df = pd.DataFrame()
thur_df = thur_df.append(df4)

df5 = pd.read_excel("xlsx/0910/221014.xlsx")
fri_df = pd.DataFrame()
fri_df = fri_df.append(df5)

# heatmap column에 넣을 리스트, 8시부터 17시만 (7시, 18시 판매 건수 제외)
timelist = mon_df.loc[0:9, '시간대'].to_list()

# 요일별로 주문횟수 리스트로 뽑기
mon_number = mon_df.loc[0:9, '주문횟수'].to_list()   
tue_number = tue_df.loc[0:9, '주문횟수'].to_list()
wed_number = wed_df.loc[0:9, '주문횟수'].to_list()
thur_number = thur_df.loc[0:9, '주문횟수'].to_list()
fri_number = fri_df.loc[0:9, '주문횟수'].to_list()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

# Defining figure size
# for the output plot
fig, ax = plt.subplots(figsize = (15, 10))


# 행에 넣어줄 인덱스 표기
idx = ['mon', 'tue', 'wed', 'thur', 'fri']

# 히트맵 넣어줄 데이터 프레임
df = pd.DataFrame([mon_number, tue_number, wed_number, thur_number, fri_number],
				columns = timelist, index = idx)


# cmap colouring ; RdYlGn, coolwarm, magma, viridis, rocket_r, cubehelix, seagreen, ice_fire, vlag, Spectral, etc.. 
# 뒤에 _r 붙이면 색깔 리버스
# https://seaborn.pydata.org/tutorial/color_palettes.html#sequential-color-palettes

sns.heatmap(df, cmap = 'Spectral_r', linewidths = 0.5, annot = True, fmt = 'd', annot_kws = {"size" : 13})
# annot을 True로 옮겨줘야 히트맵에 글씨가 쓰임, fmt포맷을 d(int)정수로
plt.title('< 1013-1019 시간별 가은 주문 고객수 히트맵 >', fontsize = 20)
plt.xlabel('시간대별', fontsize = 15)
plt.ylabel('요일별', fontsize = 15)
plt.rcParams['font.family'] = 'Malgun Gothic'
sns.set_context("poster", font_scale = 1, rc={'xtick.labelsize': 10, 'ytick.labelsize': 10})
# paper이 제일 작은 거, poster로 지정해줘야 크게 그래프가 그려짐
```
